### **Arrays**

- A numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers.

- The number of dimensions is the rank of the array;

- The shape of an array is a tuple of integers giving the size of the array along each dimension.

### **Array Operations**

```python
import numpy as np

# 建立兩個 2D 陣列 (2x3)
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[10, 20, 30],
              [40, 50, 60]])

print("陣列 A:")
print(A)

print("\n陣列 B:")
print(B)
```

```python
# Elementwise difference; both produce an array
import numpy as np
# 建立兩個 2D 陣列 (2x3)
A = np.array([[1, 2, 3],[4, 5, 6]])
B = np.array([[10, 20, 30],[40, 50, 60]])
# 方法 1：使用 - 符號
diff1 = A - B
# 方法 2：使用 numpy 的 subtract 函數
diff2 = np.subtract(A, B)
# 印出結果
print("A:")
print(A)
print("\nB:")
print(B)
print("\n逐元素相減結果（方法1 A - B）:")
print(diff1)
print("\n逐元素相減結果（方法2 np.subtract）:")
print(diff2)
```

```python
# Elementwise division; both produce an array
import numpy as np
# 建立兩個 2D 陣列 (2x3)
A = np.array([[10, 20, 30],[40, 50, 60]])
B = np.array([[2, 4, 5],[10, 10, 12]])
# 方法 1：使用 / 符號
div1 = A / B
# 方法 2：使用 numpy 的 divide 函數
div2 = np.divide(A, B)
# 印出結果
print("A:")
print(A)
print("\nB:")
print(B)
print("\n逐元素相除結果（方法1 A / B）:")
print(div1)
print("\n逐元素相除結果（方法2 np.divide）:")
print(div2)
```

```python
# Elementwise product; both produce the array
import numpy as np
# 建立兩個 2D 陣列 (2x3)
A = np.array([[1, 2, 3],[4, 5, 6]])
B = np.array([[10, 20, 30],[40, 50, 60]])
# 方法 1：使用 * 符號
prod1 = A * B
# 方法 2：使用 numpy 的 multiply 函數
prod2 = np.multiply(A, B)
# 印出結果
print("A:")
print(A)
print("\nB:")
print(B)
print("\n逐元素相乘結果（方法1 A * B）:")
print(prod1)
print("\n逐元素相乘結果（方法2 np.multiply）:")
print(prod2)
```

```python
# Dot product or matrix multiplication
import numpy as np
# A 是 2×3 矩陣
A = np.array([[1, 2, 3],[4, 5, 6]])
# B 是 3×2 矩陣
B = np.array([[7, 8],[9, 10],[11, 12]])
# 方法 1：使用 np.dot()
dot1 = np.dot(A, B)
# 方法 2：使用 @ 運算子（Python 3.5+ 支援）
dot2 = A @ B
# 印出結果
print("A:")
print(A)
print("\nB:")
print(B)
print("\n矩陣乘法結果（np.dot）:")
print(dot1)
print("\n矩陣乘法結果（@ 運算子）:")
print(dot2)
```

```python
# Denoise by filter
import cv2
import matplotlib.pyplot as plt

# 1. 讀取圖片（灰階模式）
img = cv2.imread('/content/drive/MyDrive/DLBOI.hw2/hw2.png', cv2.IMREAD_GRAYSCALE)

# 2. 檢查是否讀取成功
if img is None:
    print("❌ 無法讀取圖片，請確認路徑正確")
else:
    # 3. 旋轉 180 度
    rotated_img = cv2.rotate(img, cv2.ROTATE_180)

    # 4. 使用中值濾波去雜訊（kernel size = 5）
    denoised_img = cv2.medianBlur(rotated_img, 5)

    # 5. 顯示處理後圖片
    plt.imshow(denoised_img, cmap='gray')
    plt.axis('off')
    plt.title("Rotated 180° + Median Filter")
    plt.show()

    # 6. 儲存處理後圖片（到 Google Drive）
    save_path = '/content/drive/MyDrive/DLBOI.hw2/hw2_rotated180_denoised.png'
    cv2.imwrite(save_path, denoised_img)
    print(f"✅ 已儲存圖片到：{save_path}")
```
