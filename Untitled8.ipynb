```python
from google.colab import drive
drive.mount('/content/drive')

import time, copy, sys
import torch
import torch.nn as nn
from torch.optim import Adam
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader

# =========================
# DeviceÔºöÂÑ™ÂÖàÁî® GPU
# =========================
if not torch.cuda.is_available():
    print("‚ùå Ê≤íÊúâÂÅµÊ∏¨Âà∞ GPUÔºåË´ãÂà∞„ÄåÂü∑Ë°åÈöéÊÆµ ‚Üí ËÆäÊõ¥Âü∑Ë°åÈöéÊÆµÈ°ûÂûã„ÄçÈÅ∏ GPU ÂæåÈáçË∑ë„ÄÇ")
    sys.exit()

device = torch.device("cuda")
print("‚úÖ ‰ΩøÁî®Ë£ùÁΩÆÔºö", device, torch.cuda.get_device_name(0))

# =========================
# Dataset Ë∑ØÂæëÔºà‰Ω†ÁöÑÔºâ
# =========================
train_path = '/content/drive/MyDrive/DLBOI.hw2/train'
val_path   = '/content/drive/MyDrive/DLBOI.hw2/val'
test_path  = '/content/drive/MyDrive/DLBOI.hw2/test'

# =========================
# Transform & Dataset
# =========================
img_size = 224
imagenet_mean = [0.485, 0.456, 0.406]
imagenet_std  = [0.229, 0.224, 0.225]

train_transform = transforms.Compose([
    transforms.Resize((img_size, img_size)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(imagenet_mean, imagenet_std),
])

val_test_transform = transforms.Compose([
    transforms.Resize((img_size, img_size)),
    transforms.ToTensor(),
    transforms.Normalize(imagenet_mean, imagenet_std),
])

train_dataset = datasets.ImageFolder(train_path, transform=train_transform)
val_dataset   = datasets.ImageFolder(val_path,   transform=val_test_transform)
test_dataset  = datasets.ImageFolder(test_path,  transform=val_test_transform)

batch_size = 32

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True,  num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_dataset,   batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_dataset,  batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)

dataloaders = {'train': train_loader, 'val': val_loader}
dataset_sizes = {
    'train': len(train_dataset),
    'val': len(val_dataset),
    'test': len(test_dataset),
}
class_names = train_dataset.classes
num_classes = len(class_names)

print("Classes:", class_names)
print("Sizes:", dataset_sizes)

# =========================
# Ë®ìÁ∑¥ËàáÊ∏¨Ë©¶ÂáΩÂºè
# =========================
def train_model(model, dataloaders, criterion, optimizer, num_epochs, name):
    since = time.time()
    best_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    for epoch in range(num_epochs):
        print(f"\n[{name}] Epoch {epoch+1}/{num_epochs}")
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()
            else:
                model.eval()

            running_loss = 0.0
            running_corrects = 0

            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device, non_blocking=True)
                labels = labels.to(device, non_blocking=True)

                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                    _, preds = torch.max(outputs, 1)

                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]
            print(f"{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}")

            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_wts = copy.deepcopy(model.state_dict())

    time_used = (time.time() - since) / 60
    model.load_state_dict(best_wts)
    print(f"[{name}] ‚úÖ Best Val Acc: {best_acc:.4f}, Time: {time_used:.1f} min")

    return model, best_acc.item(), time_used

def test_model(model, dataloader, criterion, name):
    model.eval()
    running_loss = 0.0
    running_corrects = 0

    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            _, preds = torch.max(outputs, 1)
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

    loss = running_loss / dataset_sizes['test']
    acc = running_corrects.double() / dataset_sizes['test']
    print(f"[name] üß™ Test Loss: {loss:.4f} Acc: {acc:.4f}")
    return loss, acc.item()

# =========================
# Backbone ÂáΩÂºè
# =========================
def freeze_backbone(model):
    for p in model.parameters():
        p.requires_grad = False

def build_resnet18(num_classes, feature_extract=False):
    try:
        model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
    except:
        model = models.resnet18(pretrained=True)
    if feature_extract:
        freeze_backbone(model)
    in_f = model.fc.in_features
    model.fc = nn.Linear(in_f, num_classes)
    return model

def build_mobilenet_v2(num_classes, feature_extract=False):
    try:
        model = models.mobilenet_v2(weights=models.MobileNet_V2_Weights.IMAGENET1K_V1)
    except:
        model = models.mobilenet_v2(pretrained=True)
    if feature_extract:
        freeze_backbone(model)
    in_f = model.classifier[1].in_features
    model.classifier[1] = nn.Linear(in_f, num_classes)
    return model

criterion = nn.CrossEntropyLoss()

# Ë∑ëÂ∞ë‰∏ÄÈªû epochÔºåË∂≥Â§†ÂØ´Â†±Âëä
EPOCHS_FT = 5   # fine-tuning
EPOCHS_FE = 5   # feature extractor

results = {}

# =========================
# 1) ResNet-18 Fine-tuning
# =========================
res_ft = build_resnet18(num_classes, feature_extract=False).to(device)
opt_res_ft = Adam(res_ft.parameters(), lr=1e-4)
res_ft, res_ft_val, res_ft_time = train_model(
    res_ft, dataloaders, criterion, opt_res_ft,
    num_epochs=EPOCHS_FT, name="ResNet18_FT"
)
_, res_ft_test = test_model(res_ft, test_loader, criterion, "ResNet18_FT")
results["ResNet18_FT"] = (res_ft_val, res_ft_test, res_ft_time)

# =========================
# 2) ResNet-18 Feature Extractor
# =========================
res_fe = build_resnet18(num_classes, feature_extract=True).to(device)
opt_res_fe = Adam(filter(lambda p: p.requires_grad, res_fe.parameters()), lr=1e-3)
res_fe, res_fe_val, res_fe_time = train_model(
    res_fe, dataloaders, criterion, opt_res_fe,
    num_epochs=EPOCHS_FE, name="ResNet18_FE"
)
_, res_fe_test = test_model(res_fe, test_loader, criterion, "ResNet18_FE")
results["ResNet18_FE"] = (res_fe_val, res_fe_test, res_fe_time)

# =========================
# 3) MobileNetV2 Fine-tuning
# =========================
mb_ft = build_mobilenet_v2(num_classes, feature_extract=False).to(device)
opt_mb_ft = Adam(mb_ft.parameters(), lr=1e-4)
mb_ft, mb_ft_val, mb_ft_time = train_model(
    mb_ft, dataloaders, criterion, opt_mb_ft,
    num_epochs=EPOCHS_FT, name="MobileNetV2_FT"
)
_, mb_ft_test = test_model(mb_ft, test_loader, criterion, "MobileNetV2_FT")
results["MobileNetV2_FT"] = (mb_ft_val, mb_ft_test, mb_ft_time)

# =========================
# 4) MobileNetV2 Feature Extractor
# =========================
mb_fe = build_mobilenet_v2(num_classes, feature_extract=True).to(device)
opt_mb_fe = Adam(filter(lambda p: p.requires_grad, mb_fe.parameters()), lr=1e-3)
mb_fe, mb_fe_val, mb_fe_time = train_model(
    mb_fe, dataloaders, criterion, opt_mb_fe,
    num_epochs=EPOCHS_FE, name="MobileNetV2_FE"
)
_, mb_fe_test = test_model(mb_fe, test_loader, criterion, "MobileNetV2_FE")
results["MobileNetV2_FE"] = (mb_fe_val, mb_fe_test, mb_fe_time)

# =========================
# SummaryÔºöÂ†±ÂëäÁî®
# =========================
print("\n===== SUMMARY (for report) =====")
for name, (val_acc, test_acc, t) in results.items():
    print(f"{name}: Val Acc={val_acc:.4f}, Test Acc={test_acc:.4f}, Time={t:.1f} min")
```
